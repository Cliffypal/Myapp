# **Load data**

```python
 import pandas as pd
 df = pd.read_csv("https://raw.githubusercontent.com/dataprofessor/data/master/delaney_solubility_with_descriptors.csv")
 df
```

# **Data preparation**

## Data serparation x and y

```python
y = df["logS"]
y
```

```python
x = df.drop("logS" ,axis=1)
x
```

## Data spltting

```python
from sklearn.model_selection import train_test_split
x_train , x_test , y_train , y_test = train_test_split(x,y,test_size=0.2,random_state=100)
```

```python
x_train
```

```python
x_test
```

##**Model building**

### Linear Regression

**training the model**

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(x_train,y_train)
```

### **Applying the model to make pridiction**

```python
y_ir_train_pred = lr.predict(x_train)
y_ir_test_pred = lr.predict(x_test)
```

```python
 y_ir_train_pred
```

```python
 y_ir_test_pred
```

** Evaluate Model performance**

```python
from sklearn.metrics import mean_squared_error , r2_score
lr_train_mse = mean_squared_error(y_train,y_ir_train_pred)
lr_train_r2 = r2_score(y_train,y_ir_train_pred)

ir_test_mse = mean_squared_error(y_test,y_ir_test_pred)
ir_test_r2 = r2_score(y_test,y_ir_test_pred)
```

```python
print ("LR MSE (Trian): ",lr_train_mse)
print ("LR R2 (Trian): ",lr_train_r2)
print ("LR MSE (Test): ",ir_test_mse)
print ("LR R2 (Test): ",ir_test_r2)
```

```python
ir_results = pd.DataFrame (["Linear Regression",lr_train_mse,lr_train_r2,ir_test_mse,ir_test_r2]).transpose()
ir_results.columns = ["Method","Training MSE","Training R2","Test MSE","Test R2"]
ir_results
```

## Random Forest

### **Training the Modle**

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor(max_depth=2,random_state=100)
rf.fit(x_train,y_train)
```

## **Applying the Modle to make pridiction**

```python
y_rf_train_pred = rf.predict(x_train)
y_rf_test_pred = rf.predict(x_test)
```

## **Evaluate model performance**

```python
from sklearn.metrics import mean_squared_error , r2_score
rf_train_mse = mean_squared_error(y_train,y_rf_train_pred)
rf_train_r2 = r2_score(y_train,y_rf_train_pred)

rf_test_mse = mean_squared_error(y_test,y_rf_test_pred)
rf_test_r2 = r2_score(y_test,y_rf_test_pred)
```

```python
rf_results = pd.DataFrame (["Random forest",rf_train_mse,rf_train_r2,rf_test_mse,rf_test_r2]).transpose()
rf_results.columns = ["Method","Training MSE","Training R2","Test MSE","Test R2"]
rf_results
```

# **Model compararison**

```python
df_models = pd.concat([ir_results,rf_results],axis=0)
df_models.reset_index(drop=True)
```
